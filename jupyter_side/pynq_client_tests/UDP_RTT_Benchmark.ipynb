# UDP RTT Benchmark

Direct board-to-server UDP round-trip benchmark for report figures.

This path does not depend on the monitor tunnel. It sends dedicated RTT probe packets to the EC2 UDP server and measures the echo time back to the notebook.

```python
from pathlib import Path
import sys

NOTEBOOK_DIR = Path.cwd()
if (NOTEBOOK_DIR / "udp_rtt.py").exists():
    PACKAGE_PARENT = NOTEBOOK_DIR.parent
else:
    PACKAGE_PARENT = NOTEBOOK_DIR
if str(PACKAGE_PARENT) not in sys.path:
    sys.path.insert(0, str(PACKAGE_PARENT))

from pynq_client_tests.plot_udp_rtt_csv import load_rtt_rows, plot_rtt_rows, summarise_by_label
from pynq_client_tests.udp_rtt import run_udp_rtt_benchmark, _write_csv, _write_json
```

## Configure Benchmark

```python
SERVER = "3.9.71.204"
PORT = 9000
LABEL = "idle"
SAMPLES = 100
TIMEOUT_S = 1.0
CSV_PATH = Path(f"udp_rtt_{LABEL}.csv")
JSON_PATH = Path(f"udp_rtt_{LABEL}.json")

SERVER, PORT, LABEL, SAMPLES, CSV_PATH, JSON_PATH
```

## Run Benchmark

Run this while `./pynq_dev.sh` is active. No browser or monitor tunnel is needed.

```python
report, csv_rows = run_udp_rtt_benchmark(
    server=SERVER,
    port=PORT,
    label=LABEL,
    samples=SAMPLES,
    timeout_s=TIMEOUT_S,
)

_write_csv(str(CSV_PATH), csv_rows)
_write_json(str(JSON_PATH), report)

report
```
